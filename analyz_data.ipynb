```python
import pandas as pd
df = pd.read_pickle('text_train_with_answers.pkl')
```

```python
import matplotlib.pyplot as plt

label_counts = df['cheating_20%'].value_counts()

print("Распределение метки cheating_20%:")
print(label_counts)
print("\nПроцентное распределение:")
print((label_counts / len(df)) * 100)

plt.figure(figsize=(6, 6))
plt.bar(label_counts.index, label_counts.values, color=['blue', 'orange'])
plt.xticks([0, 1], ['No Cheating (0)', 'Cheating (1)'])
plt.title("Распределение метки cheating_20%")
plt.ylabel("Количество записей")
plt.xlabel("Метка")
plt.show()

plt.figure(figsize=(6, 6))
plt.pie(label_counts, labels=['No Cheating (0)', 'Cheating (1)'], autopct='%1.1f%%', colors=['blue', 'orange'])
plt.title("Процентное распределение метки cheating_20%")
plt.show()
```
